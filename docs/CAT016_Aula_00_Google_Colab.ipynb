## O que é o Google Colab?



---


Google Collaboratory ou Colab é um serviço de armazenamento em nuvem de notebooks voltados à criação e execução de códigos em Python, diretamente em um navegador, sem a necessidade de nenhum tipo de instalação de software em uma máquina.



---

Para acessá-lo são necessárias duas condições:

    

*   Um navegador web com acesso à internet;
*   Uma conta no Google.


    






## Como usar o Google Colab?



---



*   Para criar um notebook, basta acessar a página do [Colab](https://colab.research.google.com/) e, então, em **Novo notebook**
*   Um notebook será criado e estará **em branco**, pronto para receber sua primeira linha de código ou texto

*   Para dar o **nome do notebook**, basta ir ao canto superior esquerdo da tela e substituir onde está escrito Untitled0
*   Para **rodar o código**, é só clicar no pequeno ícone de play que estará no início da célula, à esquerda ou no atalho do teclado shift enter. 


*   Para incluir uma **nova célula**, clique no botão + Código no menu superior. Para incluir um texto, clique em + Texto.
*   A primeira célula de código pode demorar um pouco para executar, pois o Google está ligando uma máquina exclusivamente para você utilizá-la gratuitamente com **mais de 12 GB de RAM  e mais de 100 GB de HD**


*   No menu arquivos, tem-se várias opções como **baixar o notebook, localizá-lo no drive**, entre outras
*   Em ferramentas, pode-se por exemplo **mudar a cor da tela**, ver quais os **atalhos do teclado**, entre outros

````python
# Para cada uma das cores dentro da lista cor, eu quero printar. 
#Lembrando que tudo que está dentro do for está tabeado. 
#O print que está fora, não faz parte do loop, ou seja, ele só aparece só final, quando o loop estiver acabado
for cores in cor:
  print(cores)
  print('.')
print('acabou')
````
